````python
# ==============================================================
# DASHBOARD DE ANÁLISIS DE DISTRIBUCIÓN DE CAUDALES - FlowHydroTech
# Adaptado para df_caudal_aforos
# ==============================================================


import plotly.io as pio
pio.renderers.default = 'notebook_connected'


import pandas as pd
import numpy as np
from scipy import stats
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import ipywidgets as widgets
from ipywidgets import VBox, HBox
from IPython.display import display, HTML

# ==============================================================
# CONFIGURACIÓN DE COLOR Y DATAFRAME
# ==============================================================

COLORS = {'primary': '#2563eb', 'secondary': '#10b981', 'accent': '#f59e0b', 'danger': '#ef4444'}

_df = df_caudal_aforos.copy()

# Verificación de columnas mínimas requeridas
required_cols = ['holeid', 'fecha', 'caudal_l/s']
for c in required_cols:
    if c not in _df.columns:
        raise ValueError(f"❌ Falta la columna requerida: {c}")

_df['fecha'] = pd.to_datetime(_df['fecha'], errors='coerce')
_df = _df.dropna(subset=['fecha', 'holeid'])

print(f"✅ Dataset cargado correctamente: {len(_df):,} registros")

# ==============================================================
# WIDGETS DE SELECCIÓN
# ==============================================================

w_holeid = widgets.Dropdown(
    options=['<Todos>'] + sorted(_df['holeid'].dropna().unique().tolist()),
    description='HoleID:',
    style={'description_width': '80px'},
    layout=widgets.Layout(width='300px')
)

w_bins = widgets.IntSlider(value=30, min=10, max=100, step=5, description='Bins:')
w_remove_na = widgets.Checkbox(value=True, description='Eliminar NaN')
w_show_outliers = widgets.Checkbox(value=True, description='Mostrar outliers')

output = widgets.Output()

# ==============================================================
# FUNCIÓN DE VISUALIZACIÓN
# ==============================================================

def create_viz(*args):
    output.clear_output()
    with output:
        df_filt = _df.copy()
        if w_holeid.value != '<Todos>':
            df_filt = df_filt[df_filt['holeid'] == w_holeid.value]

        x = df_filt['caudal_l/s']
        if w_remove_na.value:
            x = x.dropna()

        if len(x) < 5:
            display(HTML("<div style='color:red; padding:20px;'>⚠️ Datos insuficientes para análisis</div>"))
            return

        # Crear figura
        fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=['📈 Distribución + KDE', '🎯 Q-Q Plot', '📦 Box Plot', '📅 Serie Temporal + Outliers'],
            specs=[[{}, {}], [{}, {}]]
        )

        # --- HISTOGRAMA + KDE ---
        fig.add_trace(go.Histogram(
            x=x, nbinsx=w_bins.value, opacity=0.7, name="Histograma",
            marker_color=COLORS['primary'], histnorm='probability density'
        ), row=1, col=1)

        if len(x) > 5:
            kde = stats.gaussian_kde(x)
            x_range = np.linspace(x.min(), x.max(), 200)
            fig.add_trace(go.Scatter(
                x=x_range, y=kde(x_range), mode='lines', name='KDE',
                line=dict(color=COLORS['secondary'], width=3)
            ), row=1, col=1)

        # --- Q-Q PLOT ---
        # Robustly coerce series to a 1D numeric numpy array before using probplot
        try:
            x_arr = np.asarray(x).ravel()
            # If dtype is object, try to coerce to numeric via pandas (handles e.g. strings)
            if x_arr.dtype == object:
                x_arr = pd.to_numeric(x_arr, errors='coerce')
            # Drop any NaNs that may have appeared during coercion
            x_arr = x_arr[~np.isnan(x_arr)]
        except Exception:
            x_arr = np.array([], dtype=float)

        # Ensure we have enough numeric values (should be true due to earlier checks)
        if x_arr.size < 5:
            # Fallback annotation if something went wrong
            fig.add_annotation(
                text="⚠️ Q-Q: datos insuficientes o no numéricos",
                x=0.5, y=0.5, xref='x domain', yref='y domain', showarrow=False,
                font=dict(size=12, color='#64748b'), bgcolor="rgba(248,250,252,0.9)",
                row=1, col=2
            )
        else:
            # Use scipy probplot on the numeric array
            try:
                prob_res = stats.probplot(x_arr, dist='norm')
                # unpack in a safe way: first element contains (osm, osr), second contains slope/intercept/r
                (osm, osr) = prob_res[0]
                slope_val, intercept_val, r_val = prob_res[1]
            except Exception:
                # As a last resort compute theoretical quantiles and sample quantiles manually
                osm = np.linspace(np.min(x_arr), np.max(x_arr), num=min(200, x_arr.size))
                osr = np.sort(x_arr)
                slope_val, intercept_val = np.polyfit(osm[:osr.size], osr, 1) if osr.size >= 2 else (1.0, 0.0)
                r_val = np.corrcoef(osr, osm[:osr.size])[0, 1] if osr.size >= 2 else float('nan')

            # Ensure arrays and numeric types
            osm = np.asarray(osm).astype(float)
            osr = np.asarray(osr).astype(float)
            try:
                slope = float(slope_val)
                intercept = float(intercept_val)
                r = float(r_val)
            except Exception:
                slope, intercept = np.polyfit(osm, osr, 1)
                try:
                    r = float(np.corrcoef(osm, osr)[0, 1])
                except Exception:
                    r = float('nan')

            fig.add_trace(go.Scatter(
                x=osm, y=osr, mode='markers', name=f'Q-Q (R²={r**2:.3f})',
                marker=dict(color=COLORS['primary'], size=6)
            ), row=1, col=2)
            fig.add_trace(go.Scatter(
                x=osm, y=(slope * osm + intercept), mode='lines', name='Referencia',
                line=dict(color=COLORS['secondary'], width=2)
            ), row=1, col=2)

        # --- BOX PLOT ---
        fig.add_trace(go.Box(
            y=x, name=w_holeid.value if w_holeid.value != '<Todos>' else 'Todos',
            boxpoints='outliers' if w_show_outliers.value else False,
            marker_color=COLORS['accent']
        ), row=2, col=1)

        # --- SERIE TEMPORAL ---
        df_temp = df_filt[['fecha', 'caudal_l/s']].dropna().sort_values('fecha')
        if len(df_temp) > 5:
            fig.add_trace(go.Scatter(
                x=df_temp['fecha'], y=df_temp['caudal_l/s'],
                mode='lines+markers',
                name='Serie Original',
                line=dict(color=COLORS['primary'], width=1.5),
                marker=dict(size=4, color=COLORS['primary']),
                hovertemplate="<b>Fecha:</b> %{x}<br><b>Caudal:</b> %{y:.3f}<extra></extra>"
            ), row=2, col=2)

            # Media móvil adaptativa
            window_size = max(3, min(30, len(df_temp)//10))
            df_temp['media_movil'] = df_temp['caudal_l/s'].rolling(window=window_size, center=True, min_periods=1).mean()
            fig.add_trace(go.Scatter(
                x=df_temp['fecha'], y=df_temp['media_movil'],
                mode='lines', name=f'Media Móvil ({window_size})',
                line=dict(color=COLORS['secondary'], width=3)
            ), row=2, col=2)

            # Outliers (IQR)
            if w_show_outliers.value:
                Q1, Q3 = x.quantile(0.25), x.quantile(0.75)
                IQR = Q3 - Q1
                lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
                df_out = df_temp[(df_temp['caudal_l/s'] < lower) | (df_temp['caudal_l/s'] > upper)]
                if len(df_out) > 0:
                    fig.add_trace(go.Scatter(
                        x=df_out['fecha'], y=df_out['caudal_l/s'],
                        mode='markers', name=f'Outliers ({len(df_out)})',
                        marker=dict(color='rgba(37,99,235,0.7)', size=10, symbol='circle',
                                    line=dict(color='#ef4444', width=2))
                    ), row=2, col=2)
                    fig.add_hline(y=upper, line_dash="dot", line_color='#f87171',
                                    opacity=0.5, annotation_text=f"Upper: {upper:.2f}", row=2, col=2)
                    fig.add_hline(y=lower, line_dash="dot", line_color='#f87171',
                                    opacity=0.5, annotation_text=f"Lower: {lower:.2f}", row=2, col=2)
        else:
            fig.add_annotation(
                text="📅 Serie temporal insuficiente",
                x=0.5, y=0.5, xref='x domain', yref='y domain', showarrow=False,
                font=dict(size=12, color='#64748b'), bgcolor="rgba(248,250,252,0.9)",
                row=2, col=2
            )

        fig.update_layout(
            height=700, width=1400,
            title=f"🔬 Análisis de Caudal - {w_holeid.value if w_holeid.value != '<Todos>' else 'Todos los Pozos'}",
            showlegend=True, template='plotly_white',
            margin=dict(l=60, r=60, t=80, b=60)
        )
        fig.show()

        # --- ESTADÍSTICOS DESCRIPTIVOS ---
        desc = x.describe()
        # Asegurar que los valores estadísticos sean floats para evitar errores de typeshed/typing
        try:
            skew_val = float(x.skew())
        except Exception:
            skew_val = float('nan')
        try:
            kurt_val = float(x.kurtosis())
        except Exception:
            kurt_val = float('nan')

        # defensiva: extraer valores del describe() como floats (por si vienen como numpy/pandas scalars)
        def _safe(val):
            try:
                return float(val)
            except Exception:
                return float('nan')

        mean_v = _safe(desc.get('mean', np.nan))
        med_v = _safe(desc.get('50%', np.nan))
        std_v = _safe(desc.get('std', np.nan))
        q1_v = _safe(desc.get('25%', np.nan))
        q3_v = _safe(desc.get('75%', np.nan))
        min_v = _safe(desc.get('min', np.nan))
        max_v = _safe(desc.get('max', np.nan))
        rango_v = max_v - min_v

        display(HTML(f"""
        <div style='background:white; padding:20px; border-radius:12px; margin-top:20px; border:1px solid #e2e8f0;'>
            <h3>📋 Estadísticos Descriptivos</h3>
            <div style='display:grid; grid-template-columns:repeat(4,1fr); gap:16px;'>
                <div><strong>Media:</strong> {mean_v:.3f}</div>
                <div><strong>Mediana:</strong> {med_v:.3f}</div>
                <div><strong>Desv.Std:</strong> {std_v:.3f}</div>
                <div><strong>Rango:</strong> {rango_v:.3f}</div>
                <div><strong>Q1:</strong> {q1_v:.3f}</div>
                <div><strong>Q3:</strong> {q3_v:.3f}</div>
                <div><strong>Asimetría:</strong> {skew_val:.3f}</div>
                <div><strong>Curtosis:</strong> {kurt_val:.3f}</div>
            </div>
            <div style='margin-top:16px; padding:12px; background:#f0f9ff; border-radius:8px;'>
                <strong>💡 Interpretación:</strong> 
                {'Distribución simétrica' if abs(skew_val) < 0.5 else 'Distribución asimétrica'} • 
                {'Forma normal' if abs(kurt_val) < 2 else 'Forma no-normal'}
            </div>
        </div>
        """))

# ==============================================================
# CONECTAR EVENTOS Y CONSTRUIR INTERFAZ
# ==============================================================

w_holeid.observe(create_viz, names='value')
w_bins.observe(create_viz, names='value')
w_remove_na.observe(create_viz, names='value')
w_show_outliers.observe(create_viz, names='value')

header = widgets.HTML("""
<div style='background:linear-gradient(135deg,#3b82f6,#1d4ed8);padding:20px;
            border-radius:12px;color:white;text-align:center;margin-bottom:20px;'>
    <h2 style='margin:0;'>💧 Análisis de Distribución de Caudales</h2>
    <p style='margin:8px 0 0 0;opacity:0.9;'>Dashboard Exploración por Pozo</p>
</div>
""")

filters = VBox([
    widgets.HTML("<h4>🔍 Filtro de Pozo</h4>"),
    HBox([w_holeid])
])

config = VBox([
    widgets.HTML("<h4>⚙️ Configuración</h4>"),
    HBox([w_bins, w_remove_na, w_show_outliers])
])

ui = VBox([header, filters, config, output],
        layout=widgets.Layout(padding='20px', background_color='#f8fafc', border_radius='16px'))

display(ui)
````

### Método de Detección de Outliers: Rango Intercuartílico (IQR)

En el análisis de distribución del dashboard, se utiliza el método **IQR (Interquartile Range)** para detectar outliers. Este método define límites inferiores y superiores a partir de los cuartiles, y considera como outliers aquellos valores que se encuentran fuera del siguiente rango:
$
\[
\text{Outliers} = \{ x\ |\ x < Q1 - 1.5 \times IQR\ \text{o}\ x > Q3 + 1.5 \times IQR \}
\]

Donde:

- **Q1** = primer cuartil (percentil 25)
- **Q3** = tercer cuartil (percentil 75)
- **IQR** = Q3 - Q1

#### Implementación en el código:

```python
Q1 = df_temp[var_name].quantile(0.25)
Q3 = df_temp[var_name].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_mask = (df_temp[var_name] < lower_bound) | (df_temp[var_name] > upper_bound)


Ventajas del método IQR:

* Es robusto ante distribuciones no normales.

* No depende de la media ni la desviación estándar.

* Reduce la sensibilidad frente a valores extremos.

* Compatible con representaciones tipo boxplot.

````python
import pandas as pd
import numpy as np

# ============================================================
# 1 CONFIGURACIÓN INICIAL
# ============================================================

df = df_caudal_aforos.copy()

# Validar columnas mínimas
required_cols = ['holeid', 'fecha', 'caudal_l/s']
for c in required_cols:
    if c not in df.columns:
        raise ValueError(f"Falta la columna requerida: {c}")

df['fecha'] = pd.to_datetime(df['fecha'], errors='coerce')
df = df.dropna(subset=['holeid', 'fecha'])

# ============================================================
# 2 ETIQUETAR OUTLIERS POR HOLEID
# ============================================================

def etiquetar_outliers(grupo):
    """Detecta y etiqueta outliers por IQR dentro de cada holeid"""
    Q1 = grupo['caudal_l/s'].quantile(0.25)
    Q3 = grupo['caudal_l/s'].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    grupo = grupo.copy()
    grupo['lower_bound'] = lower
    grupo['upper_bound'] = upper
    grupo['IQR'] = IQR
    grupo['outlier_flag'] = np.where(
        (grupo['caudal_l/s'] < lower) | (grupo['caudal_l/s'] > upper),
        'Outlier', 'Normal'
    )
    return grupo

df_etiquetado = df.groupby('holeid', group_keys=False).apply(etiquetar_outliers)

print(f"Etiquetado completado: {df_etiquetado.shape[0]:,} registros procesados.")
print(f"Pozos analizados: {df_etiquetado['holeid'].nunique()}")

# ============================================================
# 3 ESTADÍSTICOS DESCRIPTIVOS POR HOLEID
# ============================================================

stats_by_holeid = (
    df_etiquetado.groupby('holeid')['caudal_l/s']
    .agg(['count', 'mean', 'median', 'std', 'min', 'max'])
    .reset_index()
)

stats_by_holeid['skew'] = (
    df_etiquetado.groupby('holeid')['caudal_l/s'].apply(lambda x: x.skew()).values
)
stats_by_holeid['kurtosis'] = (
    df_etiquetado.groupby('holeid')['caudal_l/s'].apply(lambda x: x.kurtosis()).values
)

# ============================================================
# 4 RESUMEN DE OUTLIERS POR HOLEID
# ============================================================

resumen_outliers = (
    df_etiquetado.groupby(['holeid', 'outlier_flag'])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

# Asegurar columnas presentes
if 'Outlier' not in resumen_outliers.columns:
    resumen_outliers['Outlier'] = 0
if 'Normal' not in resumen_outliers.columns:
    resumen_outliers['Normal'] = 0

resumen_outliers['Total'] = resumen_outliers['Outlier'] + resumen_outliers['Normal']
resumen_outliers['Porcentaje_Outliers'] = (
    resumen_outliers['Outlier'] / resumen_outliers['Total'] * 100
).round(2)

resumen_outliers = resumen_outliers.sort_values('Porcentaje_Outliers', ascending=False)

print(f"Resumen de outliers generado para {len(resumen_outliers)} pozos.")

# ============================================================
# 5 EXPORTAR A EXCEL (3 hojas)
# ============================================================

output_path = "analisis_caudal_aforos_completo.xlsx"

with pd.ExcelWriter(output_path, engine='openpyxl') as writer:
    df_etiquetado.to_excel(writer, sheet_name='Datos_Etiquetados', index=False)
    stats_by_holeid.to_excel(writer, sheet_name='Estadisticos_Por_HoleID', index=False)
    resumen_outliers.to_excel(writer, sheet_name='Resumen_Outliers', index=False)

print(f"Archivo Excel exportado exitosamente: {output_path}")
````
